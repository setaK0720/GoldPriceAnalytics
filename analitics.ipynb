```python
import pandas as pd
from datetime import datetime

# pd.set_option('display.max_rows', 30)
```

```python
df = pd.read_csv('targetfile/StatementCSV_20231110_2357.csv')
df
```

```python
df['EntryTime'] = pd.to_datetime(df.OrderOpenTime)
df['CloseTime'] = pd.to_datetime(df.OrderCloseTime)
df['OrderOpenYear'] = pd.to_datetime(df['OrderOpenTime']).dt.year
df['OrderOpenMonth'] = pd.to_datetime(df['OrderOpenTime']).dt.month
df['OrderOpenDay'] = pd.to_datetime(df['OrderOpenTime']).dt.day
df['OrderOpenHour'] = pd.to_datetime(df['OrderOpenTime']).dt.hour

df['HaveTime'] = df['CloseTime'] - df['EntryTime']
df['OrderGroup'] = df['OrderComment'].str[:11]
df['Count'] = 0

groups = df.groupby(df.OrderGroup)[['OrderType', 'OrderLots', 'EntryTime', 'OrderOpenPrice', 'OrderClosePrice', 'CloseTime', 'OrderComment', 'HaveTime', 'Count']]

pips100L_df = groups.get_group("100pips_buy")
pips200L_df = groups.get_group('200pips_buy')
pips300L_df = groups.get_group('300pips_buy')
pips400L_df = groups.get_group('400pips_buy')
pips500L_df = groups.get_group('500pips_buy') 

pips100S_df = groups.get_group("100pips_sel")
pips200S_df = groups.get_group("200pips_sel") 
pips300S_df = groups.get_group("300pips_sel")
pips400S_df = groups.get_group("400pips_sel")
pips500S_df = groups.get_group("500pips_sel")
```

```python
def GetCounts(df):
    i = 0
    for index, item in df.iterrows():
        if  item['OrderLots'] == 0.01 :
            i += 1
        df.loc[index, 'Count'] = i     
        # print( item['OrderLots'])  
    return
        
GetCounts(pips500S_df)
GetCounts(pips300L_df)

pips300L_df.head(20)
```
